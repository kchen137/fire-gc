```python
import os, random, string

import numpy as np

import pandas as pd

path = os.getcwd()
```

```python
STANDARDS = pd.read_csv("StepParameters_Olson1998.txt", 
                 names=['dimer','tilt','roll','twist','shift','slide','rise'],
                 index_col="dimer", delimiter=',|={', engine="python")

STANDARDS.rise = STANDARDS.rise.str.replace('}','').astype(float)

STANDARDS
```

```python
bases = ['A','T','C','G']

seq = ''.join(random.choice(bases) for i in range(25))

print(seq)
```

```python
maindf = pd.DataFrame(columns=['#'] + STANDARDS.columns.values.tolist(),
                      index=[i for i in range(len(seq))])

maindf
```

```python
BASES = {'A':'A-T', 
         'T':'T-A', 
         'C':'C-G', 
         'G':'G-C'}
```

```python
print(seq)
for i in range(len(seq)):
    print(BASES[ seq[i] ])
```

```python
for i in range(len(seq)):
    
    maindf.at[i, '#'] = BASES[ seq[i] ]
    
    if i == 0:
        
        maindf.at[0, ['tilt','roll','twist','shift','slide','rise']] = 0
        
    else:
        
        for PAR in ['tilt','roll','twist','shift','slide','rise']:
            
            maindf.at[i, PAR] = STANDARDS.at[ seq[i-1]+seq[i] , PAR]
            
maindf
```

```python
maindf.to_csv("TEST_EROS.par")
```
